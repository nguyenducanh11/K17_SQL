```python
conn.execute("""
UPDATE student
SET course_id = (SELECT MIN(id) FROM course)
WHERE course_id IS NULL
""")
conn.commit()
```

```python
conn.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
""")
conn.commit()
```

```python
# Câu a
df_class = pd.read_sql("""
SELECT 
    class,
    COUNT(*) AS total_students,
    AVG(score) AS avg_score
FROM student
GROUP BY class
""", conn)

print(df_class)
```

```python
# Câu b
df_course = pd.read_sql("""
SELECT 
    c.course_name,
    COUNT(*) AS total_students,
    ROUND(AVG(s.score), 2) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""", conn)

print(df_course)
```

```python
# Câu c
df_rank = pd.read_sql("""
SELECT 
    c.course_name,
    COUNT(*) AS total_students,
    ROUND(AVG(s.score), 2) AS avg_score,
    CASE 
        WHEN AVG(s.score) >= 9 THEN 'Xuat sac'
        WHEN AVG(s.score) > 6 AND AVG(s.score) <= 8.9 THEN 'Tot'
        ELSE 'Kem'
    END AS xep_loai
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""", conn)

print(df_rank)
```

# Bài 3

```python
# Câu a
df_rank_student = pd.read_sql("""
SELECT 
    student_id,
    name,
    class,
    score,
    RANK() OVER (ORDER BY score DESC) AS rank
FROM student
""", conn)

print(df_rank_student)
```

```python
# Câu b
df_rank_class = pd.read_sql("""
SELECT 
    student_id,
    name,
    class,
    score,
    RANK() OVER (
        PARTITION BY class 
        ORDER BY score DESC
    ) AS rank_in_class
FROM student
""", conn)

print(df_rank_class)
```

```python
# Câu c
df_rank_course = pd.read_sql("""
SELECT 
    student_id,
    name,
    course_id,
    score,
    DENSE_RANK() OVER (
        PARTITION BY course_id
        ORDER BY score DESC
    ) AS rank_in_course
FROM student
""", conn)

print(df_rank_course)
```

```python
# top sinh viên theo điểm số 
df_all = pd.read_sql("""
SELECT *,
       DENSE_RANK() OVER (ORDER BY score DESC) AS r_high,
       DENSE_RANK() OVER (ORDER BY score ASC) AS r_low
FROM student
""", conn)

top_high_all = df_all[df_all['r_high'] <= 3]
top_low_all  = df_all[df_all['r_low'] <= 3]

print("=== TOP CAO  ===")
print(top_high_all)

print("\n=== TOP THẤP ===")
print(top_low_all)
```

```python
# top sinh viên theo điểm số theo lớp học
df_class = pd.read_sql("""
SELECT *,
       DENSE_RANK() OVER (
           PARTITION BY class ORDER BY score DESC
       ) AS r_high,
       DENSE_RANK() OVER (
           PARTITION BY class ORDER BY score ASC
       ) AS r_low
FROM student
""", conn)

top_high_class = df_class[df_class['r_high'] <= 3]
top_low_class  = df_class[df_class['r_low'] <= 3]

print("=== TOP CAO  ===")
print(top_high_all)

print("\n=== TOP THẤP ===")
print(top_low_all)
```
